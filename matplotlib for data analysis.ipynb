Display plots in matplotlib

```python
x1=np.linspace(0,10,100)
# create a plot figure
fig=plt.figure()

plt.plot(x1,np.sin(x1),'-')
plt.plot(x1,np.cos(x1),'--')
```

```python
# matplotlib.pyplot is a collection of functions that make matplotlib act like matlab not suitable for complicated cases
```

```python
# the following code produces sine and cosine curves using pyplot API
plt.figure()

# create the first of two panels and set current axis
plt.subplot(2,1,1) # (rows, columns, panel numbers)
plt.plot(x1,np.sin(x1))

# create the second of two panels and set current axis
plt.subplot(2,1,2) # (rows, columns, panel numbers)
plt.plot(x1,np.cos(x1))
```

```python
# pyplot provides state machine interface
# state machine implicitly and automatically creates figures and axes to achieve desired plot

# Example
x=np.linspace(0,2,100)

plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3,label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title('Simple Plot')

plt.legend()

plt.show()
```

```python
# formatting the style of plot
# there is a optional argument format string, indicates color and line type of plot, by default it is 'b-'

plt.plot([1,2,3,4],[1,2,9,16],'go')
plt.axis([0,6,0,20]) # 0-xmin,6-xmax,0-ymin,20-ymax # axis command specifies viewpoint of axis
plt.show
```

```python
# Working with numpy arrays
# sequences converted into numpy arrays
# below illustrates ploting several lines with different format style in onr command using arrays

# evenly sampled time at 200 intervals
t=np.arange(0.,5.,0.2)

# red dashes, blue squares, green triangles
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```

Object Oriented API

```python
#For complex , exercise more control over figures
# figure is top level container for all plot elements- contains one or more axis
# axis - represents each individual plot

# first create a grid of plots
# ax- array of two axis abjects
fig,ax= plt.subplots(2)

#call plot() method on the appropriate object
ax[0].plot(x1,np.sin(x1),'b-')
ax[1].plot(x1,np.cos(x1),'b--')
```

```python
# objective of OO-API is to have objects to apply functions and actions on, more useful when fig contains more than one subplot

fig=plt.figure() # creates reference to the figure instance

x2=np.linspace(0,5,10)
y2=x2**2

axes=fig.add_axes([0.1,0.1,0.8,0.8]) # create new axis istance using add_axes method in figure class instance fig

axes.plot(x2,y2,'r')

axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('title')
```

Bar Chart

```python
# used to compare two or more values
# plot bar chart using plt.bar

data2=[5.,25.,50.,20.]
plt.bar(range(len(data2)),data2)
plt.show()
```

Horizontal Bar Chart

```python
#plt.barh() - used to plot horizontal bar chart

data2=[5.,25.,50.,20.]
plt.barh(range(len(data2)),data2)
plt.show()
```

Error bar Chart

```python
x9=np.arange(0,4,0.2)
y9=np.exp(-x9)
e1=0.1*np.abs(np.random.rand(len(y9)))
plt.errorbar(x9,y9,yerr=e1,fmt='.-') # error distribution
plt.show();
```

Statcked bar Chart

```python
A=[15.,30.,45.,22.]
B=[15.,25.,50.,20.]
z2=range(4)
plt.bar(z2,A,color='b')
plt.bar(z2,B,color='r',bottom=A) # bottom important for staced plotting
plt.show()
```

Pie Chart

```python
# pie function used to plot pie from an array x, wedges are created proprtionally as x/sum(x).
plt.figure(figsize=(7,7))
x10=[35,25,20,20]
labels=['Computer','Electronics','Mechanical','Chemical']
plt.pie(x10,labels=labels);
plt.show()
```

Handling x and y ticks

Adding Legend

```python
x15=np.arange(1,5)
fig,ax=plt.subplots()
ax.plot(x15,x15*1.5)
ax.plot(x15,x15*3.0)
ax.plot(x15,x15/3.0)
ax.legend(['Normal','Fast','Slow']);
```

```python
x15=np.arange(1,5)
fig,ax=plt.subplots()
ax.plot(x15,x15*1.5,label='Normal')
ax.plot(x15,x15*3.0,label='Fast')
ax.plot(x15,x15/3.0,label='Slow')
ax.legend();
```
